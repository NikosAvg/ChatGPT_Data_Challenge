```python
import pandas as pd
import numpy as np
```

```python
# Read The CSV dataset
data = pd.read_csv('Datasets/customer_reviews.csv')
# Check For NaN
num_of_nan = data.isna().sum()
print(f'Number of Nans:\n{num_of_nan}')
data
```

```python
def formatDate(date: str) -> np.datetime64:
    # First split on space and keep the first part to remove the time component if it exists
    # Second split on / and re-order to match our final format YYYY-MM-DD
    date = date.split(' ')[0]
    date = date.split('/')
    date.reverse()
    date = '-'.join(date)
    return np.datetime64(date)
```

```python
# Format Dates
# As we are probably interested in the Date and not the time so much we will format all dates to YYYY-MM-DD
data['ReviewDate_formated_1'] = pd.to_datetime(data['ReviewDate'],format='mixed') # Thank our Overloard that pandas has all this helpfull functions
data['ReviewDate_formated_2'] = data['ReviewDate'].apply(formatDate) # Custom function to format the date
```

```python
data
```

```python
# Find Duplicates
num_duplicates = data.duplicated().sum()
print(f'Number of Duplicates: {num_duplicates}')
```

```python
# Clean our columns and save the dataset to a new csv
data['ReviewDate'] = data['ReviewDate_formated_1']
data = data.drop(['ReviewDate_formated_1','ReviewDate_formated_2'],axis=1)
data.to_csv('Datasets/customer_reviews_cleaned.csv',index=False)
```
