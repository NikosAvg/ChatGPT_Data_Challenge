```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
# Read The Clean CSV dataset
data = pd.read_csv('Datasets/customer_reviews_cleaned.csv')
data
```

```python
# First we are going to drop ReviewID
data.drop(['ReviewID'],axis=1,inplace=True)
data
```

```python
unique_products = len(data.ProductID.unique())
print(f'Unique Products = {unique_products}')
# Create A Histogram of th products
data.ProductID.hist(bins=unique_products)
plt.title('Number Of Products Per Category')
```

```python
# Group by 'ProductID'
grouped = data.groupby('ProductID')
grouped
```

```python
# Plot the Distribution of Ratings per Product ID
for product_id, group in grouped:
    plt.figure(figsize=(10, 6))
    plt.hist(group['Rating'], bins=5, edgecolor='black')
    plt.title(f'Histogram of Reviews for Product ID: {product_id}')
    plt.xlabel('Rating Score')
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()
```

```python
# Define a function that uses Vader to generate a Sentiment Score per review
def sentiment_scores(sentence: str)-> str:
    # Create a SentimentIntensityAnalyzer object.
    sid_obj = SentimentIntensityAnalyzer()
    # polarity_scores method of SentimentIntensityAnalyzer
    # object gives a sentiment dictionary.
    # which contains pos, neg, neu, and compound scores.
    sentiment_dict = sid_obj.polarity_scores(sentence)
    # decide sentiment as positive, negative and neutral
    if sentiment_dict['compound'] >= 0.05 :
        sent = "Positive"
 
    elif sentiment_dict['compound'] <= - 0.05 :
       sent = "Negative"
 
    else :
        sent = "Neutral"
    return sent
```

```python
# Apply the function to each of the Review To Create a sentiment column
```

```python
data['Sentiment'] = data['Review'].apply(sentiment_scores)
```

```python
data
```

```python
# Group by 'ProductID'
grouped = data.groupby('ProductID')
grouped

# Define a mapping for sentiment values to ensure consistent order
sentiment_mapping = {'Negative': 0, 'Neutral': 1, 'Positive': 2}
# Define bins to match the sentiment mapping
fixed_bins = [-0.5, 0.5, 1.5, 2.5]  
# Plot the Distribution of Ratings per Product ID
for product_id, group in grouped:
    fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))
    
    # Histogram for Rating
    axes[0].hist(group['Rating'], bins=5, edgecolor='black')
    axes[0].set_title(f'Rating Distribution for Product ID: {product_id}')
    axes[0].set_xlabel('Rating Score')
    axes[0].set_ylabel('Frequency')
    axes[0].grid(True)

    # Histogram for Sentiment
    sentiment_values = group['Sentiment'].map(sentiment_mapping)
    counts, bins, patches = axes[1].hist(sentiment_values, bins=fixed_bins, edgecolor='black')
    axes[1].set_title(f'Sentiment Distribution for Product ID: {product_id}')
    axes[1].set_xlabel('Sentiment')
    axes[1].set_ylabel('Frequency')
    axes[1].set_xticks([0, 1, 2])
    axes[1].set_xticklabels(['Negative', 'Neutral', 'Positive'])
    axes[1].grid(True)

    plt.tight_layout()
    plt.show()
    
```

```python
for product_id, group in grouped:
    group['ReviewDate'] = pd.to_datetime(group['ReviewDate'])
    # Map the sentiment to numeric values for plotting
    sentiment_mapping = {'Negative': -1, 'Neutral': 0, 'Positive': 1}
    group['Sentiment_Mapped'] = group['Sentiment'].map(sentiment_mapping)
    # Sort the group by ReviewDate
    group = group.sort_values('ReviewDate')
    fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(14, 6))
    # Plot sentiment over time
    axes[0].plot(group['ReviewDate'], group['Sentiment_Mapped'], marker='o', linestyle='-', color='blue')
    axes[0].set_title(f'Sentiment Over Time for Product ID: {product_id}')
    axes[0].set_xlabel('Date')
    axes[0].set_ylabel('Sentiment')
    axes[0].set_yticks([-1, 0, 1])
    axes[0].set_yticklabels(['Negative', 'Neutral', 'Positive'])
    axes[0].grid(True)
    # Plot rating over time
    axes[1].plot(group['ReviewDate'], group['Rating'], marker='o', linestyle='-', color='green')
    axes[1].set_title(f'Rating Over Time for Product ID: {product_id}')
    axes[1].set_xlabel('Date')
    axes[1].set_ylabel('Rating')
    axes[1].set_yticks(range(1, 6))
    axes[1].grid(True)
    plt.tight_layout()
    plt.show()
```

```python
# Mapping sentiment to numeric values
sentiment_mapping = {'Negative': -1, 'Neutral': 0, 'Positive': 1}

# Dictionary to store correlation results
correlation_results = {}

for product_id, group in grouped:
    # Map sentiment to numeric values
    group['Sentiment_Mapped'] = group['Sentiment'].map(sentiment_mapping)
    
    # Calculate Pearson correlation between Rating and Sentiment
    correlation = group['Rating'].corr(group['Sentiment_Mapped'])
    
    # Store the result for each product
    correlation_results[product_id] = correlation

# Convert the results to a DataFrame for better visualization
correlation_df = pd.DataFrame.from_dict(correlation_results, orient='index', columns=['Correlation'])
correlation_df.index.name = 'Product ID'

# Display the dataframe
print(correlation_df)

# Visualize the correlation results
plt.figure(figsize=(10, 6))
plt.bar(correlation_df.index, correlation_df['Correlation'], color='skyblue')
plt.title('Correlation Between Rating and Sentiment')
plt.xlabel('Product ID')
plt.ylabel('Correlation')
plt.grid(True)
plt.show()
```
